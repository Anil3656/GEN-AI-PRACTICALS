```python
from langchain_community.document_loaders import UnstructuredFileIOLoader

with open("python_inbuildfunctions.pdf", "rb") as f:
    loader = UnstructuredFileIOLoader(file=f)
    documents = loader.load()

documents  # This will be a list of Document objects
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size = 800, chunk_overlap = 150)
```

```python
text_splitter
```

```python
text = text_splitter.split_documents(documents)
```

```python
type(text)
```

```python
text[1]
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
persist_directory = "vector_db"
```

```python
from langchain_chroma import Chroma

vectordb = Chroma.from_documents(
    documents=text,
    embedding=embeddings,
    persist_directory=persist_directory
)
```

```python
print(vectordb)
```

```python
# Retrival
retriever = vectordb.as_retriever()
```

```python
from langchain_groq import ChatGroq
llm = ChatGroq(model = "llama-3.3-70b-versatile",temperature=0.2)
```

```python
from langchain_classic.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = 'explain about filter, map, reduce'
qa_result = qa_chain.invoke({"query":query})
```

```python
qa_result
```

```python
print(qa_result['result'])
```

```python
from langchain_core.runnables import RunnableSequence
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are an AI assistant. Use only the given context to answer.
Context:
{context}
Question:
{question}
Answer:
""")

def format_docs(docs):
    return "\n\n".join([d.page_content for d in docs])
rag_chain = (
    {"context": retriever | format_docs, "question": lambda x: x}
    | prompt
    | llm
)

response = rag_chain.invoke("Explain the PDF summary.")
response
```
